```python
import pandas as pd
import numpy as np
import seaborn as sn
import matplotlib.pyplot as plt

df_2015 = pd.read_csv("/content/drive/MyDrive/Proyectos Portafolio/Proyecto - ListenBrainz - MetaBrainz/2015.csv")
df_2016 = pd.read_csv("/content/drive/MyDrive/Proyectos Portafolio/Proyecto - ListenBrainz - MetaBrainz/2016.csv")
df_2017 = pd.read_csv("/content/drive/MyDrive/Proyectos Portafolio/Proyecto - ListenBrainz - MetaBrainz/2017.csv")
df_conteo_usuarios = pd.read_csv("/content/drive/MyDrive/Proyectos Portafolio/Proyecto - ListenBrainz - MetaBrainz/Conteo_Usuarios.csv")
df_horas_escucha = pd.read_csv("/content/drive/MyDrive/Proyectos Portafolio/Proyecto - ListenBrainz - MetaBrainz/Horas_escucha.csv")
```

```python
# Concatenado de dataframes

df_principal = pd.concat([df_2015, df_2016, df_2017], axis=0)
```

```python
df_principal.name = "df_principal"
df_conteo_usuarios.name = "df_conteo_usuarios"
df_horas_escucha.name = "df_horas_escucha"
```

```python

lista_dict = [df_principal, df_conteo_usuarios, df_horas_escucha]

for i in lista_dict:
  print(f"Dataframe: {i.name}  Type: {type(i)}  ")
  print(i.sample(5))
  print("*"*50)
```

```python
for i in lista_dict:
  print(f"Dataframe: {i.name}  Type: {type(i)}  ")
  print(i.info())
  print("*"*50)
```

```python
# Castear columnas fecha

df_principal["Fecha"] = pd.to_datetime(df_principal["Fecha"], errors= "coerce")
df_conteo_usuarios["Fecha"] = pd.to_datetime(df_conteo_usuarios["Fecha"], errors= "coerce")
df_horas_escucha['Hora'] = pd.to_datetime(df_horas_escucha['Hora'], format='%H:%M')
```

```python
for i in lista_dict:
  print(f"Dataframe: {i.name}  Type: {type(i)}  ")
  print(i.describe(include = "all"))
  print("*"*50)
```

```python
# limpieza de valores nulos

df_principal = df_principal.dropna(how="any")
df_conteo_usuarios = df_conteo_usuarios.dropna()
df_horas_escucha = df_horas_escucha.dropna()
```

```python
# Validacion de limpieza de valores nulos

for i in df_principal.columns:
  print(f"Columna: {i}")
  print(df_principal[i].isnull().sum())
  print(df_principal[i].isna().sum())
  print("*"*50)

for i in df_conteo_usuarios.columns:
  print(f"Columna: {i}")
  print(df_conteo_usuarios[i].isnull().sum())
  print(df_conteo_usuarios[i].isna().sum())
  print("*"*50)


for i in df_horas_escucha.columns:
  print(f"Columna: {i}")
  print(df_horas_escucha[i].isnull().sum())
  print(df_horas_escucha[i].isna().sum())
  print("*"*50)
```

```python
import re

df_artist_filtrado  = df_principal[df_principal['artist_name'].str.contains(r'^(?!\d+$)[a-zA-Z0-9]+$', regex = True)]
df_artist_filtrado.head(5)
```

```python

grupo_df_principal = df_artist_filtrado.groupby(["Fecha","artist_name","track_name"])["Conteo"].sum()
grupo_df_principal = grupo_df_principal.reset_index()
df_principal = pd.DataFrame(grupo_df_principal)
df_principal
```
